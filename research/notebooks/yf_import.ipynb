```python
from clients.yahoo_finance_client import YahooFinanceClient
from ticker_mapping import TICKER_MAPPING
from datetime import datetime, timedelta, timezone
import requests
```

```python
client = YahooFinanceClient()
tickers = [info['ticker'] for info in TICKER_MAPPING.values()]
```

```python
len(tickers)
```

```python
%%capture

now_utc = datetime.now(timezone.utc).replace(second=0, microsecond=0) - timedelta(minutes=1)
bootstrap_start = now_utc - timedelta(days=3)
df = client.download_ohlc_list(tickers, "1m", bootstrap_start, now_utc)
```

```python
numner_of_symbols = df['ticker'].unique().shape[0]
print("Number of unique symbols downloaded:", numner_of_symbols)
```
